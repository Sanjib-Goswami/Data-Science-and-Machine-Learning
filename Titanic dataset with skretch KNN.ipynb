```python
x = pd.read_csv('xdata.csv')
y = pd.read_csv('ydata.csv')

x = x.drop('0', axis=1).values
y = y.drop('0',axis=1).values
```

```python
x
```

```python
y
```

```python
def distance(a,b):
    dist = np.sqrt(sum(a-b)**2)
    return dist
```

```python
def KNN(x,y,querry, k=7):
    m = x.shape[0]
    n = querry.shape[0]
    predict=[]
    for j in range(n):
        dist=[]
        for i in range(m):
            d = distance(x[i], querry[j])
            dist.append([d,y[i][0]])            ## here y is 2D...
        sort_dist = sorted(dist)
        sort_dist = sort_dist[:7]
        sort_dist = np.array(sort_dist)
        no_of_label = np.unique(sort_dist[:,1], return_counts=True)
        index = no_of_label[1].argmax()
        predict.append([querry[j], no_of_label[0][index]])
    return predict    
```

```python
x_train = x[0:int(x.shape[0]*0.7)]
x_test = x[int(x.shape[0]*0.7):]

y_train = y[0:int(x.shape[0]*0.7)]
y_test = y[int(x.shape[0]*0.7):]
```

```python
x_train.shape , y_train.shape , x_test.shape , y_test.shape
```

```python
predict = KNN(x_train,y_train,x_test)
```

```python
predict
```

```python
b = np.unique(y_train, return_counts=True)
b
```

```python
def score(x,y):
    count=0
    for i in range(len(y)):
        if x[i][1]==y[i]:
            count=count+1
    print(count)
    print(count/len(y))
    print('\n')
    print(count/len(y)*100)
    
```

```python
score(predict, y_test)
```

```python
train = pd.read_csv('titanic_train.csv')
test = pd.read_csv('titanic_test.csv')
test.head()
```

```python
train.drop(['PassengerId','Name','Ticket','Cabin'], axis=1, inplace=True)
train.head(n=10)
```

```python
train = train.fillna({"Age":train["Age"].mean()})
train
```

```python
train.isnull().sum()
```

```python
train.dropna(inplace=True)
train.head()
```

```python
## Labeling in the sex column......
l=[]
def convert_sex(sex):
    #print(sex)
    for i in sex:
        if i=='male':
            l.append(1)
        else:
            l.append(0)
convert_sex(train["Sex"])
train["Sex"]=l
train
```

```python
## Labeling the Embarked column....
ls=[]
def convert_em(embarked):
   # print(embarked)
    for i in embarked:
        if i=="S":
            ls.append(1)
        elif i=="C":
            ls.append(2)
        else:
            ls.append(3)
convert_em(train['Embarked'])
train["Embarked"]=ls
train
```
