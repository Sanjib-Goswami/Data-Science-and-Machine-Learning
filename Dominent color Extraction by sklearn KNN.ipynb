```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Natural.jpg')                 ## it reads all pixels from the image in BGR form..
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)      ## convert bgr to rgb..
```

```python
img.shape
```

```python
img
```

```python
plt.imshow(img)
plt.show()
```

```python
all_pixel = img.reshape(-1,3)
all_pixel
```

```python
all_pixel.shape
```

```python
from sklearn.cluster import KMeans
dominant_color=4
km = KMeans(n_clusters=dominant_color)     ## by how many color we want show the image....
```

```python
km.fit(all_pixel)
```

```python
center=km.cluster_centers_
print(center)
```

```python
center = np.array(center , dtype="uint8")
center
```

```python
km.labels_
```

```python
set(km.labels_)
```

```python
np.unique(km.labels_ , return_counts=True)
```

```python
np.unique(km.labels_)
```

```python
new_img = np.zeros((all_pixel.shape[0],3))
```

```python
new_img.shape
```

```python
for i in range(new_img.shape[0]):
    new_img[i] = center[km.labels_[i]]
```

```python
new_img = new_img.reshape((img.shape[0],img.shape[1],3))
new_img = np.array(new_img , dtype="uint8")               ## convert all the pixel from float to unsinged integer...
```

```python
plt.imshow(new_img)
plt.show()
```
